## Ensemble Retriever

### Loading the dataset

```python
from langchain_community.document_loaders import HuggingFaceDatasetLoader

dataset_name = "MedRAG/textbooks"
page_content_column = "content"

loader = HuggingFaceDatasetLoader(dataset_name, page_content_column)
data = loader.load()
```

### Defining retrievers

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.retrievers import BM25Retriever, EnsembleRetriever
from langchain.docstore import InMemoryDocstore
from langchain.schema import Document
from tqdm import tqdm
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2", model_kwargs={"device": device})

splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=50)
doc_splits = splitter.split_documents(data)

bm25_retriever = BM25Retriever.from_documents(doc_splits)

batch_size = 1000
batches = [doc_splits[i:i + batch_size] for i in range(0, len(doc_splits), batch_size)]
```

```python
vector_store = Chroma(collection_name="rag-chroma", embedding_function=embeddings, persist_directory="./chroma_db") # changed from chromadb to langchian-chroma, in case of any errors, change the pip install

for batch in tqdm(batches, desc="Adding batches to Chroma vector store"):
    vector_store.add_documents(batch)


vector_retriever = vector_store.as_retriever()
```

```python
retriever = EnsembleRetriever(retrievers=[bm25_retriever, vector_retriever])

query = "What is pneumonia?"
results = retriever.invoke(query)

for i, doc in enumerate(results):
    print(f"Result {i + 1}: {doc.page_content}")
```

## LLM setup

```python
import google.generativeai as genai

config = {
    "temperature" : 0.9,
    "top_p" : 1,
    "top_k" : 1,
    "max_output_tokens" : 10
}

class GeminiChat:
    def __init__(self, api_key=None, model_name='gemini-pro'):
        if api_key:
            genai.configure(api_key=api_key)
        self.model = genai.GenerativeModel(model_name,generation_config=config)

    def generate(self, prompt):
        """Generate text based on the provided prompt."""

        if type(prompt) != str :
          # print("Received prompt: ",prompt)
          prompt = prompt.messages[-1].content
        response = self.model.generate_content(prompt)
        return response.text

llm = GeminiChat(
    api_key="",
    model_name="gemini-pro"
)

prompt = "The opposite of hot is"
response = llm.generate(prompt)
print("Generated Response:", response)
```

## Workflow

```python
from typing import Sequence, Literal
from pydantic import BaseModel, Field
from langchain import PromptTemplate, hub
from langchain_core.messages import BaseMessage, HumanMessage
from dataclasses import dataclass, field

@dataclass
class AgentState:
    response: BaseMessage = None
    query_asked: BaseMessage = None
    retrieved_docs: Sequence[BaseMessage] = field(default_factory=list)  # contains any number of chunks retrieved from the vector store
    retrieve_loop_counter: int = 0  # Counter for rewrites
    hallucination_loop_counter: int = 0  # Counter for hallucination checks
    decision: Literal["generate", "rewrite", "Output"] = None
    query_context: str = ""

rag_prompt = hub.pull("rlm/rag-prompt")
rag_prompt.messages[0].prompt.template = """You are an assistant for question-answering tasks. Use the following pieces of retrieved context to answer the question, if needed. Provide a concise answer not exceeding more than 4 words. No other output except your answer is acceptable
Question: {question}
Question context: {query_context}
Context: {context}
Answer:"""

def retrieve_documents(state: AgentState):  # HOW DOES IT TAKE INPUT?
    """
    Custom retrieval logic for documents.
    Maps the AgentState to the expected input format for retriever_tool.
    """
    print("---RETRIEVING DOCUMENTS---")
    query = state.query_asked.content  # Extract the user query
    input_data = {"query": query}  # Map to expected format for retriever_tool
    retrieved_docs = retriever_tool.invoke(input_data)  # Pass the correct input
    state.retrieved_docs.append(HumanMessage(content=retrieved_docs))  # Add retrieved docs to state NEEDS WORK!!!

    return state

def grade_documents(state: AgentState):
    """
    Determines whether the retrieved documents are relevant to the question.
    """
    print("---CHECK RELEVANCE---")
    prompt = PromptTemplate(
        template="""You are a grader assessing relevance of a retrieved document to a user question.
        Here is the retrieved document: \n\n {context} \n\n
        Here is the user question: {question} \n
        If the document contains keyword(s) or semantic meaning related to the user question, grade it as relevant.
        Give a binary score 'yes' or 'no'.""",
        input_variables=["context", "question"],
    )

    question = state.query_asked.content
    docs = state.retrieved_docs[-1].content  # Last message is assumed to be the retrieved docs  NEEDS WORK
    model = llm
    combined_prompt = prompt.format(context=docs, question=question)
    score = model.generate(combined_prompt).strip().lower()

    decision = None

    if score == "yes":
        print("---DECISION: DOCS RELEVANT---")
        decision = "generate"
    else:
        print("---DECISION: DOCS NOT RELEVANT---")
        decision = "rewrite"
        state.retrieve_loop_counter += 1

        if state.retrieve_loop_counter >= 3:
            print("---REWRITE LIMIT REACHED: FORCING GENERATION---")
            decision = "generate"

    return AgentState(
        response = HumanMessage(content=decision),
        query_asked = state.query_asked,
        retrieved_docs = state.retrieved_docs,
        retrieve_loop_counter=state.retrieve_loop_counter,
        hallucination_loop_counter=state.hallucination_loop_counter,
        decision=decision
    )

def rewrite(state: AgentState) -> AgentState:
    """
    Transform the query to produce a better question.
    """
    print("---TRANSFORM QUERY---")
    question = state.query_asked.content
    model = llm

    response = model.generate(
        f"""You are provided a question as the input. Understand the underlying semantic intent/meaning and improve the question which can facilitate better responses.
        Here is the initial question: {question}. Formulate an improved question. You must respond with just the modified query in plain text."""
    )

    state.query_asked = HumanMessage(content=response)
    state.decision = None

    return state

def generate(state: AgentState) -> AgentState:
    """
    Generate an answer using the retrieved documents and the user query.
    """
    print("---GENERATE---")
    question = state.query_asked.content
    docs = state.retrieved_docs[-1].content  # SHOULD BE CHANGED

    model = llm
    rag_chain = rag_prompt | model.generate
    response = rag_chain.invoke({"context": docs, "question": question,"query_context": state.query_context})

    state.response = HumanMessage(content=response)
    state.decision = None
    return state

def check_hallucinations(state: AgentState):
    """
    Checks if the generated response appropriately answers the user query.
    """
    print("---CHECK HALLUCINATIONS---")
    prompt = PromptTemplate(
        template="""You are a verifier assessing whether the generated response answers the user question appropriately.
        Here is the generated response: {response} \n\n
        Here is the user question: {question} \n
        Respond with 'yes' if the response appropriately answers the question, otherwise respond with 'no'.""",
        input_variables=["response", "question"],
    )

    question = state.query_asked.content
    response = state.response.content

    model = llm
    combined_prompt = prompt.format(response=response, question=question)
    score = model.generate(combined_prompt).strip().lower()

    decision = None

    if score == "yes":
        print("---DECISION: RESPONSE APPROPRIATE---")
        decision = "Output"

    else:
        state.hallucination_loop_counter += 1
        decision = "rewrite"
        print("---DECISION: RESPONSE NOT APPROPRIATE---")

        if state.hallucination_loop_counter >= 3:
            print("---HALLUCINATION CHECK LIMIT REACHED: FORCING OUTPUT---")
            decision = "Output" # Force output after 3 retries

    return AgentState(
        response = HumanMessage(content=response),
        query_asked = state.query_asked,
        retrieved_docs = state.retrieved_docs,
        retrieve_loop_counter=state.retrieve_loop_counter,
        hallucination_loop_counter=state.hallucination_loop_counter,
        decision=decision
    )
```

```python
from IPython.display import display, Image
from langgraph.graph import END, StateGraph, START
from langgraph.utils.runnable import RunnableCallable
from typing import Annotated, Sequence


def routing_function(state:AgentState):
  return state.decision


workflow_graph = StateGraph(state_schema=AgentState)

workflow_graph.add_node("Retrieve Documents", retrieve_documents)
workflow_graph.add_node("Check Relevance", grade_documents)
workflow_graph.add_node("Rewrite Query", rewrite)
workflow_graph.add_node("LLM", generate)
workflow_graph.add_node("Check Hallucinations", check_hallucinations)

workflow_graph.add_edge(START, "Retrieve Documents")
workflow_graph.add_edge("Retrieve Documents", "Check Relevance")

workflow_graph.add_conditional_edges("Check Relevance", routing_function, {"generate": "LLM", "rewrite": "Rewrite Query"})

workflow_graph.add_edge("LLM", "Check Hallucinations")
workflow_graph.add_edge("Rewrite Query", "Retrieve Documents")

workflow_graph.add_conditional_edges("Check Hallucinations", routing_function, {"Output": END , "rewrite": "LLM"})


compiled_graph = workflow_graph.compile()

try:
    graph_image = compiled_graph.get_graph(xray=True).draw_mermaid_png()
    display(Image(graph_image))
except Exception as e:
    print(f"Error displaying the graph: {e}")
```

```python
from rouge_score import rouge_scorer

def calculate_total_rouge(true_labels, predicted_labels):

    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    combined_true = " ".join(true_labels)
    combined_predicted = " ".join(predicted_labels)

    total_score = scorer.score(combined_true, combined_predicted)

    return total_score

total_rouge_scores = calculate_total_rouge(true_labels, predicted_labels)
print("Total ROUGE Scores:",total_rouge_scores)
```

```python
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction

def calculate_bleu(true_labels, predicted_labels):

    references = []
    hypotheses = []
    for true, pred in zip(true_labels, predicted_labels):
        references.append([true.split()])
        if pred.strip():
            hypotheses.append(pred.split())
        else:
            hypotheses.append([])

    smoothing_function = SmoothingFunction().method1
    bleu_score = corpus_bleu(references, hypotheses, smoothing_function=smoothing_function)
    return bleu_score



bleu_score = calculate_bleu(true_labels, predicted_labels)
print("BLEU Score:", round(bleu_score,3))
```

```python
from sentence_transformers import SentenceTransformer, util
def calculate_llm_similarity(true_labels, predicted_labels, model_name="all-MiniLM-L6-v2"):
    model = SentenceTransformer(model_name)

    similarities = []
    for true, pred in zip(true_labels, predicted_labels):
        if pred:
            embeddings = model.encode([true, pred], convert_to_tensor=True)
            similarity = util.pytorch_cos_sim(embeddings[0], embeddings[1]).item()
            similarities.append(similarity)
        else:
            similarities.append(0.0)

    return similarities


similarities = calculate_llm_similarity(true_labels, predicted_labels)
round(sum(similarities)/len(similarities),3)

```
